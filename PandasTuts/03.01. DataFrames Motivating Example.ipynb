## Importing the Necessary Packages

```python
import pandas as pd
import numpy as np
```

#### The Next example is a data frame generated from the standard normal distribution using numpy and pandas.

Also, this example shows some Pandas incredible functionalities such as:

  - Shape
  
  - info()
  
  - describe()
  
  - head()
  
  - dtypes

```python
# Creating index values
ind = ["row" + "_" + str(num) for num in range(1, 16)]

# Creating column names (Variables)
cols = ["var" + "_" + str(num) for num in range(1, 7)]

# Generating Random Data
data = np.random.randn(90).reshape(15, 6)
```

```python
# Constructing a Data Frame 
df = pd.DataFrame(data = data, index = ind, columns = cols)
```

```python
df.head()
```

```python
# Get Some Information about the data
df.info()
```

```python
df.dtypes
```

```python
# Basic Summary Statistics
df.describe()
```

```python
import seaborn as sns
%matplotlib inline
```

```python
sns.pairplot(df[['var_1', 'var_2', 'var_3','var_4']], diag_kind = 'kde')
```

```python
## We might need to check the available methods in Pandas DataFrames
import pprint as p
p.pprint([meth for meth in dir(pd.DataFrame) if '__' not in meth], indent= 4,
         width = 60, compact=True)
```
