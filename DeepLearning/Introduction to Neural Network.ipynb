## Why Neural Network

 - Assume a regression problem, the model predicts the target value in an additive way by a multiply-add manner. 
 
 - Regression assumes no interaction among features wich is rare in the real world, because features interact with each other. For example:
     - When one feature goes up another feature may go down.
     - When two features go up or down another feature may go up or down as well. 
     
- In fact, we don't know which feature interact with each other.

- **The main reason for Neural Network is  to account for `interactions and relationships among features`.**

## Feed Forward Propagation

- Forward propagation simply means moving from left to right, in other words:
   - From Input Layer ----> Output layer
   
   - The layers are connected through lines (we call them **weights**)
   
   - The Input layer feeds its content to the next layer.
   
   - The output of a layer will be an input to the next layer.

#### Hidden Layer Nodes:    
    
$$hidden\_node\_1 = (2 * 1) + (3 * 1) = 2 + 3 = 5$$

$$hidden\_node\_2 = (2 * (-1)) + (3 * 1) = -2 + 3 = 1$$
    
#### The Output Layer Node

$$output\_node = (5 * 2) + (1 * (-1)) = 10 - 1 = 9$$

```python
##------------------------------------------------------------------
#           Manually forward propagation
##------------------------------------------------------------------
import numpy as np

input_layer = np.array([2, 3])
weights = {"node_0": np.array([1, 1]),
           "node_1": np.array([-1, 1]),
           "output": np.array([2, -1])}

# Calculate node 0 value: node_0_value

node_0_value = (input_layer * weights['node_0']).sum()

# Calculate node 1 value: node_1_value
node_1_value = (input_layer * weights['node_1']).sum()

# Put node values into array: hidden_layer_outputs
hidden_layer_outputs = np.array([node_0_value, node_1_value])

# Calculate output: output
output = (hidden_layer_outputs * weights['output']).sum()

print("="*40)
print("01- The value of hidden layer Node One is: " + str(node_0_value))
print("02- The value of hidden layer Node Two is: " + str(node_1_value))
print("="*40)
print("03- The value of the output layer Node is: " + str(output))
print("="*40)
```

## Two-Hidden Layer Neural Network with Two Nodes Example 01

## Manual Calculation

#### Hidden Layer 1 Node Values:    
    
$$hidden\_node\_1\_1 = (4 * 2) + (3 * -1) = 5$$

$$hidden\_node\_1\_2 = (4 * 1) + (3 * -1) = 1$$

    
#### The Hidden Layer 2 Node Values:

$$hidden\_node\_2\_1 =(5 * 3) + (1 * -2) = 13$$

$$hidden\_node\_2\_2 = (5 * 4) + (1 * 6) = 26$$


#### The Output Layer  Node Value:

$$Output\_node= (13 * -5) + (26 * 4) = 39$$

## Python Solution

```python
import numpy as np
from numpy import array

## Create the Input Layer Data
input_data = np.array([4, 3])

## Create a dictionary of The Weights
weights = {'node_0_1': array([2, -1]),
           'node_0_2': array([1, -1]),
           'node_1_1': array([3, -2]),
           'node_1_2': array([4, 6]),
           'output_node': array([-5, 4])}

## Compute the node values
node_0_1_output = (input_data * weights['node_0_1']).sum()
node_0_2_output = (input_data * weights['node_0_2']).sum()

print("="*40)
print("The First Hidden Layer Node values are: ")
print("="*40)
print(node_0_1_output)
print(node_0_2_output)
print("="*40)

## Convert the value into a vector
#===================================
hidden_outputs_1 = np.array([node_0_1_output, node_0_2_output])

print("The First Hidden Layer Output is: ")
print("="*40)
print(hidden_outputs_1)
print("="*40)

hidden_output_1_1 = (hidden_outputs_1 * weights['node_1_1']).sum()
hidden_output_1_2 = (hidden_outputs_1 * weights['node_1_2']).sum()


print("The Second Hidden Layer Node values are: ")
print("="*40)
print(hidden_output_1_1)
print(hidden_output_1_2)


## Convert the value into a vector
#===================================
hidden_outputs_2 = np.array([hidden_output_1_1, hidden_output_1_2])


print("="*40)
print("The Second Hidden Layer Output is: ")
print("="*40)
print(hidden_outputs_2)

output_node = (hidden_outputs_2 * weights['output_node']).sum()

print("="*40)
print("The Output Layer value is: ")
print("="*40)
print(output_node)
print("="*40)
```

#### Hidden Layer 1 Node Values:    
    
$$hidden\_node\_1\_1 = (5 * 2) + (7 * 3) + (3 * 4)= 43$$

$$hidden\_node\_1\_2 = (5 * 5) + (7 * -1) + (3 * -2)= 12$$

#### The Hidden Layer 2 Node Values:

$$hidden\_node\_2\_1 =(43 * -1) + (12 * 1) + (7 * 6)= 11$$

$$hidden\_node\_2\_2 = (43 * -1) + (12 * 2) + (7 * 6)= 23$$

#### The Output Layer  Node Value:

$$Output\_node= (11 * -2) + (23 * 2) + (0 * 6)= 24$$

### Solving a Two_Hidden Layer Example with Python

```python
from numpy import array

## Create the Input Layer Data
input_data = np.array([5, 7, 3])

## Create a dictionary of The Weights
weights = {'node_0_1': array([2, 3, 4]),
           'node_0_2': array([5, -1, -2]),
           'node_0_3': array([-2, 2, 1]),
           'node_1_1': array([-1, 1, 6]),
           'node_1_2': array([-1, 2, 6]),
           'node_1_3': array([-2, -3, 6]),
           'output_node': array([-2, 2, 6])}

## Compute the node values
node_0_1_output = (input_data * weights['node_0_1']).sum()
node_0_2_output = (input_data * weights['node_0_2']).sum()
node_0_3_output = (input_data * weights['node_0_3']).sum()

print("="*40)
print("The First Hidden Layer Node values are: ")
print("="*40)
print(node_0_1_output)
print(node_0_2_output)
print(node_0_3_output)
print("="*40)


## Convert the value into a vector
#===================================
hidden_outputs_1 = np.array([node_0_1_output, node_0_2_output, node_0_3_output])

print("The First Hidden Layer Output is: ")
print("="*40)
print(hidden_outputs_1)
print("="*40)

hidden_output_1_1 = (hidden_outputs_1 * weights['node_1_1']).sum()
hidden_output_1_2 = (hidden_outputs_1 * weights['node_1_2']).sum()
hidden_output_1_3 = (hidden_outputs_1 * weights['node_1_3']).sum()


print("The Second Hidden Layer Node values are: ")
print("="*40)
print(hidden_output_1_1)
print(hidden_output_1_2)
print(hidden_output_1_3)

## Convert the value into a vector
#===================================
hidden_outputs_2 = np.array([hidden_output_1_1, hidden_output_1_2, hidden_output_1_3])

print("The Second Hidden Layer Output is: ")
print("="*40)
print(hidden_outputs_2)

output_node = (hidden_outputs_2 * weights['output_node']).sum()

print("="*40)
print("The Output Layer value is: ")
print(output_node)

# We update the last value of the second hidden layer (-80) and put it as (0)
update_second_hidden_layer = array([11, 23, 0])

updated_output_node = (update_second_hidden_layer * weights['output_node']).sum()

print("="*40)
print("The updated second Layer vector is: ")
print("="*40)
print(update_second_hidden_layer)


print("="*40)
print("The Correct Output Layer value is: ")
print("="*40)
print(updated_output_node)
print("="*40)
```

# Activation Functions

**What is activation function?**


- **Activation function** is a simple function that transforms its inputs into outputs that have a certain range. 

**Why do we need an activation function in neural network?**

1. We know that complex patterns are non-linear. Thus, we need a way to account for that.

2. Capture non-linearity in the model to help learn the complex patterns in the data. 

3. If we don't use an activation function, then neural network won't be any difference linear function. 


### Popular Activation Functions

  2. Sigmoid function
  3. Linear Function
  4. Softmax
  5. ReLU (Reclified Linear Function)
  6. ELU (Exponontial Linear Function)
  7. Leaky-ReLU

## The Rectified Linear Activation Function

- The rectified linear activation function (called ReLU) has been shown to lead to very high-performance networks. This function takes a single number as an input, returning 0 if the input is negative, and the input if the input is positive.

### The math formula of ReLU function

### \begin{equation}
 ReLU(x) =
    \begin{cases}
      x & {\text{if }} x>0 \\
      0 & {\text{otherwise}}
    \end{cases}       
\end{equation}

Or simply:

### $$f(x) = max(0,x)$$

- Here are some examples:
    - relu(4) = 3
    - relu(-7) = 0

```python
def ReLU(x):
    output = []
    for i in x:
        if i > 0:
            output.append(i)
        else:
            output.append(0)
    return(output)
```

```python
from matplotlib import pyplot
import numpy as np

x = list(range(-20, 20))
y = ReLU(x)

pyplot.style.use('Solarize_Light2')
pyplot.figure(figsize=(10,5))

pyplot.plot(x, y, c = "red")
pyplot.title("ReLU Activation Function", fontweight="bold")
pyplot.show()
```

## ReLU Acctivation Function Example

   - To practice many examples, it is better to write a function. And that is what the next function will do. 

```python
from numpy import array

def relu(input):
    '''
    The ReLU acctivation function
    '''
    output = max(0, input)
    return(output)

def two_hidden_layer_network(input_data, weights):
    """
    This function predicts neural networks with:
    one input layer with three nodes
    Two Hidden layers
    
    This function uses ReLU acctivation function
    """
    
    # The First Hidden Layer
    node_0_1_input = (input_data * weights['node_0_1']).sum()
    node_0_1_output = relu(node_0_1_input)

    node_0_2_input = (input_data * weights['node_0_2']).sum()
    node_0_2_output = relu(node_0_2_input)
    
    node_0_3_input = (input_data * weights['node_0_3']).sum()
    node_0_3_output = relu(node_0_3_input)

    hidden_0_outputs = np.array([node_0_1_output, node_0_2_output, node_0_3_output])
    
    # The Second Hidden Layer
    node_1_1_input = (hidden_0_outputs * weights['node_1_1']).sum()
    node_1_1_output = relu(node_1_1_input)

    node_1_2_input = (hidden_0_outputs * weights['node_1_2']).sum()
    node_1_2_output = relu(node_1_2_input)
    
    node_1_3_input = (hidden_0_outputs * weights['node_1_3']).sum()
    node_1_3_output = relu(node_1_3_input)

    hidden_1_outputs = np.array([node_1_1_output, node_1_2_output, node_1_3_output])
    

    # The Output Layer
    output = (hidden_1_outputs * weights['output']).sum()
    model_output = relu(output)
    
    # Print the results to see what is happening inside the network
    print("="*75)
    print(hidden_0_outputs)
    print(hidden_1_outputs)
    #print(hidden_2_outputs)
    print(model_output)
    
    return(model_output) 
```

```python
two_hidden_layer_network(input_data1, weights1)
```
