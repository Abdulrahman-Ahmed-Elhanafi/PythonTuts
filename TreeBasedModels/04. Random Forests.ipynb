## Introduction:

  - **Random forests** or **random decision forests** are an **ensemble learning method** for classification, regression and other tasks that operates by constructing a **large number of decision trees** at training time. In other words, the random forest algorithm builds multiple decision trees and aggregate them to get an accurate result.
  
  
  - For classification tasks, the output of the random forest is the class selected by most trees, while for regression tasks, the mean or average prediction of the individual trees is returned.
  
  
  - All the decision trees that make up a random forest are different because we build each tree on a different random subset of our data.
  
  
   - A random forest tends to be more accurate than a single decision tree because it minimizes overfitting.
  
  
  - Random forests generally outperform decision trees, but their accuracy is lower than **gradient boosted trees** (which we will see in a later lecture).


### Diversity with Random Forests
    
   - Random Forests is an ensemble method that uses a decision tree as a **base estimator**.
   
   
   - **Random Forest** is a classifier that contains a number of decision trees on various subsets of the given dataset and takes the average to improve the predictive accuracy of that dataset." Instead of relying on one decision tree, the random forest takes the prediction from each tree and based on the **majority votes** of predictions, and it predicts the final output.
   
   
   - The greater number of trees in the forest leads to higher accuracy and prevents the problem of overfitting.
   
   
   - In Random Forests, each estimator is trained on a different bootstrap sample having the **same size** as the training set. 
   
   
   - Random forests introduces further randomization than **bagging** when training each of the base estimators. When each tree is trained, only **d** features can be sampled at each node without replacement, where **d** is a number smaller than the total number of features.
   
   
### Note: 


- Recall, in **bagging** the **base estimator** could be any model including a decision tree, logistic regression or even a neural network. Each estimator is trained on a distinct bootstrap sample drawn from the training set using all available features.

### The Steps of building a Random Forest

 - Each tree in a random forest is built using the following steps where A represents the entire forest, a represents a single tree, for a = 1 to A:
    1. Create a bootstrap sample with replacement, D training from $x$, $y$ label these $X_a$, $Y_a$
    2. Train the tree fa on $X_a$, $Y_a$
    3. Average the predictions or take the majority vote to arrive at a final prediction
    
In a regression problem, predictions for the test instances are made by taking the mean of the predictions made by all trees. This can be represented as follows:

### $$\hat p = \frac{1}{N}\sum_{a=1}^{A}f_a(x)$$

**Where**, 
 
 - **N**: the total number of trees
 
 - **a=1**: The first tree in the forest
 
 - **$f_a(x)$** represents the prediction from each tree

### Random Forests Training Workflow

 - The training procedure for random forests works as follows:
     - Each tree forming the ensemble is trained on a different bootstrap sample from the training set. 
     - When a tree is trained, at each node, only **d** features are sampled from all features without replacement.
     - The node is then split using the sampled feature that maximizes **information gain**.
     

#### Note:
- In scikit-learn **d** defaults to the square-root of the number of features. For example, if there are 100 features, only 10 features are sampled at each node.

___

### Random Forests: Prediction

- Once trained, predictions can be made on new instances. When a new instance is fed to the different base estimators, each of them outputs a prediction. The predictions are then collected by the random forests meta-classifier and a final prediction is made depending on the nature of the problem.

 - For classification, the final prediction is made by majority voting. 
 
 - For regression, the final prediction is the average of all the labels predicted by the base estimators.

## <center><font size=6, color="#7B249F"><u> Random Forests Implementation in sklearn</u> </font>

The scikit-learn library implements random forests by providing two estimators:

   - `RandomForestClassifier` for classification problems
   
   - `RandomForestRegressor` for regression problems. 

**The options (parameters) of random forests estimators**: there are several parameters can be passed to the estimators mentioned above, we will mention the mose common ones:

   - **n_estimators**: the number of trees the algorithm builds before making predictions. the higher the number the better the performance but it costs more in terms of computation.
   
   
   - max_features: the maximum number of features that the random forest is allowed to try in an individual tree.
   
   
   - **min_sample_leaf**: the minimum number of leaves that are required to split an internal node.
   
   
   - **n_jobs**: how many jobs to run in parallel. A value of -1 means it will use all the processors.
   
   
   - **random_state**: to produce the same results when using the algorithm on the same training dataset.
   
   
   - **oob_score** or **out-of-the-bag sampling**: This is a random forest cross-validation method. In this sampling method, about one-third of the data is not used to train the model and can be used to evaluate its performance. These samples are called the **out-of-the-bag samples**.
 

### Training a Random Forest Regressor:

Here we mention the steps of training a random forest resgressor:
  - **Step 01**: Import the regressor `RandomForestRegressor` from `sklearn.ensemble`
  
  - **Step 02**: Import the splitting function `train_test_split` from `sklearn.model_selection`, also import the evaluation metric `mean_squared_error` from `sklearn.metrics`  
  - **Step 03**: Split the data into train and test sets
  - **Step 04**: Instantiate the regressors and set:
      - **Number of estimators**: using `n_estimators` parameter
      - **random_state** to have reproducible results using `random_state` parameter.
      - **min_sample_leaf** and other parameters based on your needs
  - **Step 05**: Fit the regressor on the training set
  - **Step 06**: Generate predictions
  - **Step 07**: Assess the model performance on test set
  
    
**Python Syntax for train a random forest regressor**,

```python
# Import random forest algorithm
# 1. For regression
from sklearn.ensemble import RandomForestRegressor

# 2. For classification
from sklearn.ensemble import RandomForestClassifier

# Import analysis tools

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE

# Split dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=0)

 # Instantiate a random forests classifier
rf = RandomForestClassifier(n_estimators=100,
                           min_samples_leaf=0.13,
                           random_state=0)

# Fit 'rf' to the training set
rf.fit(X_train, y_train)
# Predict the test set labels 'y_pred'
y_pred = rf.predict(X_test)
                           
# Evaluate the test set RMSE
rmse_test = MSE(y_test, y_pred)**(1/2)
# Print the test set RMSE
print('Test set RMSE of rf: {:.2f}'.format(rmse_test))                         
```

## Auto Dataset Project Using Random Forest Regressor

````python
## ======================================================================
#            Importing the necessary modules and tools
## ======================================================================

import pandas as pd; import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

# import DecisionTreeRegressor from sklearn.tree 
from sklearn.ensemble import RandomForestRegressor

# from sklearn.metrics import necessary metrics
from sklearn.metrics import mean_squared_error as MSE


# Set notebook options
# --------------------
pd.options.display.float_format = '{:,.3f}'.format
%matplotlib inline
````

````python
auto = pd.read_csv('auto_mpg.csv')
print("*" * 50)
auto.info()
print("*" * 50)
auto.head()
````

````python
auto['horsepower'] = pd.to_numeric(auto.horsepower, errors = 'coerce')
auto.dropna(inplace=True)

# Mapping the origin variable to names
# ------------------------------------

auto['origin'] = auto['origin'].map({1: 'USA',
                                     2: 'Europe',
                                     3: 'Japan'})
# Encoding the origin variable
# ------------------------------
auto = pd.get_dummies(auto, columns = ['origin'], prefix = '', 
                      prefix_sep = '', drop_first=True)
X = auto.drop(['mpg', 'car name'], axis=1)
y = auto['mpg']

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y,
                                                    test_size=0.20, 
                                                    random_state=3)
````

````python
# Instantiate a random forests regressor 'rf_reg'
# ----------------------------------------------
rf_reg = RandomForestRegressor(random_state= 1)

# Fit the tree regressor
# ----------------------
rf_reg.fit(X_train, y_train)

# Generate predictions
# ---------------------
preds = rf_reg.predict(X_test)

# Compute the rmse
rmse = np.sqrt(MSE(preds, y_test))


# Print the rmse metric
# -----------------------
print("Test set RMSE: {:.2f}".format(rmse))
````

## <center><font size=6, color="#7B249F"><u> Feature Importance</u> </font>

- When a tree based method is trained, the predictive power of a feature or its importance can be assessed. 
    - In `scikit-learn`, feature importance is assessed by measuring how much the tree nodes use a particular feature to reduce impurity.
    - The importance of a feature is expressed as a percentage indicating the weight of that feature in training and prediction.
    - Once a tree-based model is trained in `scikit-learn`, the features importances can be accessed by extracting the `feature_importance_` attribute from the model.

### Feature Importance Visualization


## Random Forest Hyperparameter Tuning using Grid Seach CV

  - To illustrate this technique, we will use the same example used in Regression trees. And we perform grid-search cross-validation on the auto-dataset.
  
  - The steps are straightforward, as we are becoming familiar with them day by day, no there is no harm to mention them again:
  
**Step 01**: import all the tools you need for analysis such as `mean_squared_error` from `sklearn.metrics` and `GridSearchCV` from `sklearn.model_selection`.

**Step 02**:  Define a dictionary containing the grid of hyperparameters. 

**Step 03**:  instantiate a GridSearchCV object and:

   - pass the regressor to estimator parameter.
   
   - dictionary of hyperparameters to param_grid
   
   - Set the number of folds **cv** (5 or 10 ...)
   
   - Set scoring metric, for example `neg_mean_squared_error` in case of regression problem as in our case. in order to use negative mean squared error as a metric. 
   
   - Verbose parameter controls verbosity; the higher its value, the more messages are printed during fitting. (This is useful to see whether you are going to wait fot the results or leaving to do something else).

**Step 04**:  Fitting the grid seach to the training data.

**Step 05**:  Extracting the best results:

 - The best hyperparameters using `.best_params_` attribute.
 
 - The best estimator using `.best_estimator_` attribute.
 
 - The best score using `.best_score_` attribute.
You can extract rf's best hyperparameters by getting the attribute  from grid_rf. The results are shown here.

**Step 06**: Evaluating the best model performance:

  - You can save the best model from the regressor using  `.best_estimator_` then predict the results on the test set. 
  
  - Assess the test set metric (RMSE). 


**Hyperparameter Tuning Syntax**:

### Hyperparameter Tuning for Auto Dataset Project

````python
## ==========================================================
#   Random Forest Regressor Hyperparameter tuning In Action
## ==========================================================

# Import GridSearchCV
#----------------------
from sklearn.model_selection import GridSearchCV

# Instantiate a tree regressor
# ----------------------------
rf_reg = RandomForestRegressor(random_state= 1)

# Define the dictionary 'params_rf_reg'
params_rf_reg = {
    'n_estimators': [100, 350, 500], 
    'max_features': ['log2', 'auto', 'sqrt'],
    'min_samples_leaf': [2, 10, 30]
}

# Import GridSearchCV
# ------------------------
from sklearn.model_selection import GridSearchCV

# Instantiate grid_rf_reg
grid_rf_reg = GridSearchCV(estimator=rf_reg,
                       param_grid= params_rf_reg,
                       scoring='neg_mean_squared_error',
                       cv=5,
                       verbose=1,
                       n_jobs=8)

 # Fit 'grid_rf_reg' to the training set
grid_rf_reg.fit(X_train, y_train)

# Extract the best estimator
best_model = grid_rf_reg.best_estimator_

# Predict test set labels
y_pred = best_model.predict(X_test)

# Compute rmse_test
rmse_test = np.sqrt(MSE(y_test, y_pred))

# Print rmse_test
print('Test RMSE of best model: {:.3f}'.format(rmse_test)) 
````
