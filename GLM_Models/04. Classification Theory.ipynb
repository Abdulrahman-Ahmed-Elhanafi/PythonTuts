# <center> <u><font color="#7B241C"> Classification Theory and Model Performance Metrics </u> 

## Classification

### Supervised Classification

  - In supervised classification, the target is a class label. 


  - Classification is a task that requires the use of machine learning algorithms that learn how to assign a class label to examples from the problem domain. In other words, a predictive algorithm assigns, to each class, a score that measures the propensity that the case belongs to a particular class. 


 - In statistics, classification is the problem of identifying which of a set of categories (sub-populations) an observation, (or observations) belongs to. [source](https://en.wikipedia.org/wiki/Statistical_classification)

 - An algorithm that implements classification, is known as a classifier.
 
### Examples of Classification:

   - Classifying emails as “spam” or “not spam.”
   
   - Classifying user behavior, "churn" or "not churn".
   

### Types of Classification 


  - **Binary Classification**: is the task of classifying the elements of a set into two groups on the basis of a classification rule. In other words, it refers to predicting one of two classes.
  
  
  - **Multiclass Classification**: involves predicting one of more than two classes. In machine learning, multiclass or multinomial classification is the problem of classifying instances into one of three or more classes. 
  

# Classification Metrics For Model Performance

# Confusion Matrix 

## Understanding the Confusion Matrix

- Typically, predictive models are used to categorize cases using a **cutoff (or decision boundary)**. For example:
     1. Building model to predict a disease; so,  if the probability of having a disease is above the cutoff, further diagnosis and analyses will be carried out. If it's below the cutoff, then the person is not sick. 
     2. Prdicting fraud, if the probability is above the cutoff, then you need to investigate the operation, otherwise the operation is not fraud. 
     
### Model Assessment
     
   - The fundamental assessment tool for classification model performance is the **confusion matrix**.
   - Classification models are evaluated for each given **cutoff**. For example, you choose a cutoff then you calculate the confusion matrix. You repeat this process many times until you find the best model. Later, we will see another tool for determining the **optimal cutoff**.

### Confusion Matrix Definition

 - The confusion matrix is a cross tabulation of **predicted classes and actual classes**. It quantifies the confusion of the classifier. 
 - Creating a confusion matrix requires a cutoff. 

### The Structure of Confusion Matrix

- Before we dive into the structure of confusion matrix we need to master some terminologies:
    - **Target Event**: takes the value of:
        - **1**, meaning the event did occur and it is usually **rare**, and it can be:
            - **Favorable**: like responding to an offer.
            - **Unfavorable**: like a financial transaction is fraud, having a disease, churning, plane crashing ... etc.
    - **Non-event**: takes the value of **0**, meaning the event did not occur.
            
            
  - Whether the target event is favorable or unfavorable, it is called **positive** when classified as **1**.
  - The target event is called **negative** when it is classified as 0.
   
   
- **Actual Class**: Binary classes:
    - **Actual negative**: an actual class of 0, or when the event did not occur.
    - **Actual positive**: an actual class of 1, or when the event did occur.
    
    
- **Predicted Class**: there are two cases in binary classification:
    - **Predicted negative**: a predicted class of 0, or predict an event did not occur.
    - **Predicted negative**: a predicted class of 1, or predict an event did occur.


- **Confusion Matrix components**: There are four cases:
    - **True Negative (TN)**: the target is **negative and predicted negative**. This happens when the probability **below the cutoff**.
    - **True Positive (TP)**: The target is **positive and predicted positive**. This happens when the probability **above the cutoff**.
    - **False Positive (FP)**: the target is **actually negative but we predict it as positive**, in other words, we predict the event will occur while it does not. This happens when the probability **above the cutoff**.
    - **False Negative (FN)**: the target is **actually positive but we predict it as negative**, in other words, we predict the event will not occur while it does. This happens when the probability **below the cutoff**.
    
### False Positives and False Negatives Examples 
 
  1. **False Positives**:
      - Predict that a man **smokes** while he **he does not**.
      - Predict a client will **churn** while he **wouldn't**.
      - Predict a borrower will **default** while **he won't**.
      - Predict a patient **having a disease** while he is **healthy**.
      - Predict a plane **will crash** while it is in a **good condition**.
  2. **False Negatives**:
      - Predict that a man **does not smoke** while he **he does**.
      - Predict a client **will not churn** while he **would**.
      - Predict a borrower will **not default** while **he will**.
      - Predict a patient **not having a disease** while he is **sick**.
      - Predict a plane **will not crash** while it is in a **bad condition**.
      
  
### Confusion Matrix Example:

- Suppose we are building a model for detecting **Fraud**:

    1. If the actual class is 0 (not fraud), and the model predicts class 0 (not fraud), then the model classified the case **correctly** **(True Negative)**.
    2. If the actual class is 0 (not fraud), and the model predicts class 1 (fraud), then the model missclassified the case **(False positive)**.
    3. If the actual class is 1 (fraud), and the model predicts class 0 (not fraud), then the model missclassified the case **(False Negative)**.
    4. If the actual class is 1 (fraud), and the model predicts class 1 (Fraud), then the model classified the case **correctly** **(True Positive)**.

## Classification Metrics

### Accuracy

   $$\textbf{Accuracy} = \frac{\text{True Positives + True Negatives}}{\text{Total number of cases}}$$
   
### or

$${\displaystyle \mathrm {ACC} ={\frac {\mathrm {TP} +\mathrm {TN} }{\mathrm {P} +\mathrm {N} }}={\frac {\mathrm {TP} +\mathrm {TN} }{\mathrm {TP} +\mathrm {TN} +\mathrm {FP} +\mathrm {FN} }}}$$

### Error Rate

$$\textbf{Error Rate} = \frac{\text{False Positives + False Negatives}}{\text{Total number of cases}} = 1 - \textbf{Accuracy}$$

### or

$$\textbf{Error Rate} ={\frac {\mathrm {FP} +\mathrm {FN} }{\mathrm {TP} +\mathrm {TN} +\mathrm {FP} +\mathrm {FN}}} = 1 - \textbf{Accuracy}$$


### The Drawbacks of Looking only to Accuracy


 - In Machine Learning, the balance between events and non-events matters; so, if the number of cases of events (which is usually rare) is less than the number of non-events, we say that our sample is **biased**. 
 
 
 - Building a machine learning model on a biased dataset will lead to biased model, this is a reason for looking at other metrics other than the overall accuracy when evaluating a machine learning algorithm.



### Sensitivity (True Positive Rate)

  - **Sensitivity**: is a metric that measures the ability of the algorithm to predict a __positive__ outcome when the **actual** outcome is __positive__. 
  
$$\textbf {Sensitivity}= \frac{\textbf{True Positives}}{\textbf{Total number of actual Positives}}$$

### Or


$$\textbf {Sensitivity}= \frac{TP}{(TP + FN)}$$
  
### Positive Predictive Value (PPV)

$$\textbf {PPV}= \frac{\textbf{True Positives}}{\textbf{Total Number of Predicted Positives}}$$

### Or

$$\textbf {PPV} = \frac{TP}{(TP + FP)}$$

 - If the algorithm predicts everything __positive__, $TN=0$ then $\textbf{sensitivity} = 1$ which is called perfect sensitivity.
 
 
### Specificity (False Positive Rate)

  - **Specificity**: is a metric that measures the ability of the algorithm to predict a __negative__ outcome when the **actual** outcome is __negative__. 

$$\textbf {Specificity}= \frac{\textbf{True Negatives}}{\textbf{Total number of actual Negatives}}$$


### Or

$$\textbf {Specificity}= \frac{TN}{(TN + FP)}$$


### Negative Predictive Value (NPV)

$$\textbf {NPV}= \frac{\textbf{True Negatives}}{\textbf{Total Number of Predicted Negatives}}$$

### Or

$$\textbf {NPV} = \frac{TN}{(TN + FN)}$$

### Balanced Accuracy:

 - **Balanced Accuracy** is another metric to summarize the specificity and sensitivity into one single metric. This is because it is preferred to have one metric rather to evaluates two metrics.
 
 $$\textbf {Balanced Accuracy}= \frac{TPR + TNR}{2}$$


### F1 Score

- **F1-Score** is a one-number metric used to evaluate a machine learning algorithm, and it is as **the harmonic mean of two rates, precision (PPV) and sensitivity (recall: TPR)**:

$$F_{1} = 2 \times \frac {PPV \times TPR}{PPV + TPR} = \frac {2 TP}{2 TP + FP + FN}$$

For more information about classfication metric check the [appendix](#Appendix:-Classification-Metrics)

## Which Classification Model Peroformance Metric

- You might be wondering which of these measures will be the most useful to you. The answer depends on the context of your problem. For example, 
    - if you are working on a target marketing project, you might be most concerned with getting a high positive predicted value because you want to maximize the response rate for customers who receive an offer. 
    - If you are working on a fraud detection project. In this case, you might be most concerned with obtaining high sensitivity such that the model is detecting a high proportion of the actual fraud cases.
    - The case of plane safety, it is important to maximize sensititivity over specificity, bacause failing to predict a plane will malfunction before it crashes is more costly error than grounding a plane when it is actually in good condition.
    - In capital murder criminal case, maximizing specificity is prefered, since false positive can lead to executing an innocent person.

# <u>The Reciever Operative Characteristic Curve (ROC)</u>



## Measuring Performance Across Cutoffs by Using the ROC Curve

- Knowing the metrics of model performance measures, like sensitivity and specificity, is just the first step. However, these metrics are calculated using a specified cutoff. The question that pops up is; **what if we don't know which cutoff to use?** 

- Another tools is available to help us choose the **optimal cutoff** by measuring the model performance across a range of cutoffs values. This tools is called **Receiver-Operating Characteristic Curve، or ROC curve**. In fact, the ROC curve was adapted from signal detection theory for the assessment of classifiers. 

- The ROC curve (figure above) displays the **sensitivity (True Positive rate)** on the Y-axis, and **1 minus specificity (False Positive Rate)** on the X-axis  for the **entire range of cutoffs**.

### ROC Curve Structure

 - The ROC curve has **sensitivity** on the **Y axis**, and **1 minus specificity** on the X axis. 
 
 - The ROC curve displays the **sensitivity and 1 minus specificity** for the **entire range of cutoff values**.
 
- **The Y axis**: This axis represents ${\color{blue}{\textbf{Sensitivity (True Positive Rate)}}}$:
 - **We know that sensitivity equals the true positives divided by the total actual positives**. 
     - Suppose you have ${\color{red}{\textbf{100 actual positives}}}$:
     
         - If you have ${\color{red}{\textbf{zero true positives}}}$, then you have ${\color{darkgreen}{\textbf{a sensitivity of 0}}}$.
         
         - If ${\color{red}{\textbf{all 100 of the actual positives are also true positives}}}$, then you have ${\color{darkgreen}{\textbf{a sensitivity of 1}}}$. 

${\textbf{This is why the Y axis goes from 0 to 1}}$.


- **The X axis**: This axis represents ${\color{blue}{\textbf{1 minus the specificity (False Positive Rate)}}}$:

 - We know that specificity equals the **true negatives divided by the total actual negatives**. 
     - Suppose you have ${\color{red}{\textbf{100 actual negatives}}}$:
     
         - If you have ${\color{red}{\textbf{zero true negatives}}}$, then you have a ${\color{darkgreen}{\textbf{specificity of 0}}}$.
         
             - $\textbf{One minus a specificity of 0 equals 1}$, so a ${\color{darkgreen}{\textbf{specificity of 0}}}$ is **located at 1 on the X axis**.
             
         - If ${\color{red}{\textbf{all 100 of the actual negatives are also true negatives}}}$, then you have ${\color{darkgreen}{\textbf{a specificity of 1}}}$. 
         
             - $\textbf{One minus a specificity of 1 equals 0}$, so a ${\color{darkgreen}{\textbf{specificity of 1}}}$ is **located at 0 on the X axis**. 
             

${\textbf{This is why the X axis goes from 0 to 1.}}$
         
## ROC Curve Readings 

- We know that predicted probabilities are constrained to be between 0 and 1, so the range of all possible cutoff values is between 0 and 1.

### ROC Curve Scenarios


 1. **A cutoff of 0**: In this scenario, all cases will have a predicted probability above the cutoff so:
 
       - All cases are in predicted class 1.
       - the sensitivity is 1, because all of the actual positives are also true positives.
       - Specificity is 0, because there are zero true negatives.
       - One minus specificity is one (False positive rate is 1)
       - The lowest possible cutoff of 0 is located at coordinate (1,1) (The upper right corner). 
       

2. **A cutoff of 1**: in this scenario all cases will have a predicted probability below the cutoff so:

      - All cases are in predicted class 0. 
      - The sensitivity is 0, because there are zero true positives.
      - The specificity is 1, because all of the actual negatives are also true negatives.
      - One minus a specificity of 1 is 0, (False positive rate is 0).
      - The highest possible cutoff of 1 is located at coordinate (0,0) (The lower left corner). 


3. **As the cutoff decreases**:

    - More cases are allocated to class 1.
    - The sensitivity increases.
    - The specificity decreases.


4. **As the cutoff increases**:
    - More cases are allocated to class 0.
    - The sensitivity decreases.
    - The specificity increases.


5. If the posterior probabilities were arbitrarily assigned to the cases:
    - The ratio of false positives to true positives would be the same as the ratio of the total actual negatives to the total actual positives. 


6. **The baseline**:
    - Displaying a 45-degree angle from (0,0) to (1,1).
    - The baseline represents a random model for which the area under the curve is 0.50, which is represented by the `C-statistic`(also known as concordance statistic, and it ranges from 0.5 to 1). 
    - This baseline represents the accuracy of a model that predicts classes no better than a random chance. 


7. The more the ROC curve bulges from the 45-degree line, the more accurately the model predicts compared to a random model.


8. If your model had predicted everything correctly (perfect prediction), then its ROC curve would reach the (0,1) point.


9. The more the ROC curve bulges toward (0,1) (the upper left corner) the better the model.

# The Area Under Curve

- ROC curve does not give us a precise metric, wouldn't be better if we have one . If fact there is a metric that is extracted from the ROC curve, the **Area Under curve or AUC** metric[1](https://en.wikipedia.org/wiki/Receiver_operating_characteristic#Area_under_the_curve). 

- The larger the **AUC** the better the model.

- The **AUC** ranges from 0.5 (random classifier) to 1 (perfect classifier)

- If AUC equals to 1, means the model classifies everything positive (True positive rate is 1 and false positive rate of zero) (The upper left corner of ROC curve).

## The Area Under Curve with Python

**Syntax**

```python
## Import the auc score 
from sklearn.metrics import roc_auc_score

## Import logistic classifier 
from sklearn.linear_model import LogisticRegression

## Import train_test_split
from sklearn.model_selection import train_test_split

## Instantiate the classifier
logreg = LogisticRegression()

## Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                   test_size = 0.25, random_state = 10123)

## Fit the model
logreg.fit(X_train, y_train)

## Predict the probability
y_pred_prob = logreg.predict_proba(X_test)[:, 1]

## Calculate Area Under Curve
roc_auc_score(y_test, y_pred_prob)
```

# Gain Curve (Gain Chart)

- Until now, we evaluated models using the AUC. Although this measure is very useful for data scientists, it is less appropriate if you want to discuss your models with business managers or non-technical people. 

- The **AUC** is a rather complex evaluation measure that is not that intuitive. Moreover, it is a single number, that does not catch all the information about the model. Instead, one can use evaluation curves like the **cumulative gains curve**. This type of curve is easy to explain and can guide you to better business decisions.

## Cumulative gains construction


- The cumulative gains curve is constructed as follows:
    - First, you order all the observations according to the output of the model. On the left hand side are observations with the highest probability to be target according to the model, on the right hand side are observations with the lowest probability to be target.
    - On the horizontal axis of the cumulative gains curve, it is indicated which percentage of the observations is considered. For instance, at 30%, the 30% observations with the highest probability to be target is considered.
    - On the vertical axis, the curve indicates which percentage of all targets is included in this group. For instance, if the cumulative gains is 70% at 30%, it means that when taking the top 30% obervations with highest probability to be target, this group contains already 70% of all targets. [source](https://www.ibm.com/docs/en/spss-statistics/24.0.0?topic=overtraining-cumulative-gains-lift-charts)

## Cumulative gains interpretation

- The cumulative gains curve is a great tool to compare models:
    - The more the line is situated to the upper left corner, the better the model.

- It is often the case that two models produce curves that cross each other. In that case, it is not straightforward to decide which model is best. For instance, (see next figure) you could say that model 2 is better to distinguish the top 10% observations from the rest, while model 1 is better to distinghuish the top 70% of the observations from the rest.

# Lift Curve

- A lift chart is a variation of the gains chart. 

- The lift equals the **positive predicted value divided by the marginal rate**.

$${\textbf{Lift}} = \frac{\textbf{Positive Predictive Value}}{\pi_1}$$

Where: $\mathcal \pi_1$ is the **marginal rate**.

- For a given depth, there are lift times more responders targeted by the model than by random chance. For example, suppose you want to target the top 10% of cases, and your lift chart shows that you will get a lift of 4. That means that when you target the top 10%, you get four times more responders than if you took a random sample of 10% of the cases. Just like with a regular gains chart, a lift chart for a model with good predictive power has a curve shaped like a steep ski slope. This shape shows that cases with the highest predicted probabilities have the highest lift. 

- You can use either a gains chart or a lift chart to show how well your model generalizes.

A plot of sensitivity versus depth is sometimes called a Lorenz curve, concentration curve, or a lift curve (although the lift value is not explicitly displayed). [souce](http://www2.cs.uregina.ca/~dbd/cs831/notes/lift_chart/lift_chart.html)
